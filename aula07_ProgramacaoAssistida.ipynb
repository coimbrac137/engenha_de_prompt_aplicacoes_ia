Soma Lista

```
# Cria uma lista com números inteiros
numeros = [1, 3, 6, 7, 8]

# Define uma variável para guardar o total da soma
soma = 0

# Percorre cada valor dentro da lista 'numeros'
for num in numeros:
  # Soma o valor atual à variável 'soma'
  soma += num

# Exibe o resultado final da soma de todos os valores
print(soma)
```

Eh primo


```
def eh_primo(n):
    # Se for menor que 2, não é primo
    if n < 2:
        return False

    # Testa se existe algum número que divide n
    for i in range(2, int(n**0.5) + 1):

        # Se dividir certinho, não é primo
        if n % i == 0:
            return False

    # Se nenhum número dividiu, é primo
    return True


# Aqui pedimos um número para o usuário
numero = int(input("Digite um número: "))

# Chamamos a função passando o número
resultado = eh_primo(numero)

# Mostramos o resultado
if resultado:
    print("É primo")
else:
    print("Não é primo")
```

Inversão de String


```
def inverte_texto(texto):
    # Cria uma lista vazia para guardar as letras
    letras = []

    # Percorre o texto original
    for letra in texto:
        # Adiciona cada letra na lista
        letras.append(letra)

    # Inverte a lista de letras
    letras.reverse()

    # Junta tudo em uma única string novamente
    return ''.join(letras)


# Pede um texto para o usuário
texto_usuario = input("Digite um texto: ")

# Chama a função passando o texto
resultado = inverte_texto(texto_usuario)

# Mostra o resultado
print("Texto invertido:", resultado)
```

#Calculadora Simples

com duas Operaçoes "+" e "-"

```
# Função de soma
def somar(a, b):
    return a + b

# Função de subtração
def subtrair(a, b):
    return a - b

# Função de multiplicação
def multiplicar(a, b):
    return a * b

# Função de divisão
def dividir(a, b):
    return a / b


# Pede os números para o usuário
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

# Pede a operação
operacao = input("Digite a operação (+, -, * ou /): ")


# Verifica qual operação o usuário quer
if operacao == "+":
    resultado = somar(num1, num2)
    print("Resultado:", resultado)

elif operacao == "-":
    resultado = subtrair(num1, num2)
    print("Resultado:", resultado)

elif operacao == "*":
    resultado = multiplicar(num1, num2)
    print("Resultado:", resultado)

elif operacao == "/":
    resultado = dividir(num1, num2)
    print("Resultado:", resultado)

else:
    # Caso o usuário digite algo errado
    print("Operação inválida!")
```
